```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.cm import coolwarm
from matplotlib.cm import Wistia
from matplotlib.colors import Normalize

texas_shapefile_path = "~/Documents/ECE Next/TexasSHP/State.shp"
texas_map = gpd.read_file(texas_shapefile_path)

desired_date = "2016-09-28"
desired_hour = '1:00:00 AM'
# AM = 오전 | PM = 오후
korean = '오전 8:00:00'
filtered_df = merged_df[(merged_df['Date'].astype(str) == desired_date) & (merged_df['Time'].astype(str) == desired_hour)]
weather_filter_df = temp_2016_df[(temp_2016_df['Date'].astype(str) == desired_date) & (temp_2016_df['Time'].astype(str) == korean)]


first_row_volts = filtered_df.iloc[0]
voltage_values = []

for volt in first_row_volts[3:]:
    if isinstance(volt, float):
        voltage_values.append(volt)

first_row_temp = weather_filter_df.iloc[0]
temp_values = []

for temp in first_row_temp[9:]:
    if temp != "No Value":
        temp = int(temp)
        if isinstance(temp, int):
            temp_values.append(temp)
print(min(temp_values))
print(max(temp_values))

volt_dict = first_row_volts.to_dict()
temp_dict = first_row_temp.to_dict()

#print(voltage_values)
volt_norm = Normalize(0.99, 1.16) # HARD CODED
temp_norm = Normalize(40, 93)

fig, ax = plt.subplots(figsize=(10, 10))
texas_map.plot(ax=ax, color='darkgray', edgecolor='gray')
weather_set = set()

# Plot voltage data first
for index, row in buses_df.iterrows():
    bus_ID_with_PU= str(row['Number']) + " PU Volt"
    bus_ID = int(row['Number'])
    ws_correlated_row= correlations_df[correlations_df['Bus ID'] == bus_ID]
    ws = ws_correlated_row['Weather Station Name (ICAO)'].iloc[0]
    ws_cord_row = ws_cords[ws_cords['Name'] == ws]
    ws_latitude = ws_cord_row['Latitude'].iloc[0]
    ws_longitude = ws_cord_row['Longitude'].iloc[0]
    
    bus_latitude = row['Substation Latitude']
    bus_longitude = row['Substation Longitude']
    voltage = volt_dict.get(bus_ID_with_PU, 0.99)
    volt_color = coolwarm(volt_norm(voltage))
    ax.plot(bus_longitude, bus_latitude, marker='.', color=volt_color, markersize=4)

# Plot temperature data last
for index, row in buses_df.iterrows():
    bus_ID_with_PU= str(row['Number']) + " PU Volt"
    bus_ID = int(row['Number'])
    ws_correlated_row= correlations_df[correlations_df['Bus ID'] == bus_ID]
    ws = ws_correlated_row['Weather Station Name (ICAO)'].iloc[0]
    ws_cord_row = ws_cords[ws_cords['Name'] == ws]
    ws_latitude = ws_cord_row['Latitude'].iloc[0]
    ws_longitude = ws_cord_row['Longitude'].iloc[0]
    
    if ws not in weather_set:
        weather_set.add(ws)
        temp = temp_dict.get(ws, 70)
        if temp != "No Value":
            temp = int(temp)
        else:
            temp = 70
        temp_color = Wistia(temp_norm(temp))
        ax.plot(ws_longitude, ws_latitude, marker='o', color=temp_color, markersize=5, markeredgecolor='black', markeredgewidth=0.5)

# Add color bar
volt_sm = plt.cm.ScalarMappable(cmap=coolwarm, norm=volt_norm)
volt_sm.set_array(voltage_values)

temp_sm = plt.cm.ScalarMappable(cmap=Wistia, norm=temp_norm)
temp_sm.set_array(temp_values)


cbar_volt= plt.colorbar(volt_sm, ax=ax, shrink=0.8)
cbar_volt.set_label('Voltage')

cbar_temp = plt.colorbar(temp_sm, ax=ax, shrink=0.8)
cbar_temp.set_label('Temperature (F)')

ax.set_title(f"Date: {desired_date} | Hour: {desired_hour}")
plt.show()
```

```python
import numpy as np
date_power = '9/11/2007'
date_weather = '2007-09-11'

date_df = abn_vol_times_df[abn_vol_times_df['Date'] == date_weather]
hours = list(date_df['Time'])

convert_hours = hours.copy()
hours_korean = []
for hour in convert_hours:
    if 'PM' in hour:
        hour = hour.replace('PM', '오후')
    if 'AM' in hour:
        hour = hour.replace('AM', '오전')
    last_three_characters = hour[-3:]
    rest_of_hour = hour[:-3]
    last_three_characters = last_three_characters[-2:]
    last_three_characters = last_three_characters + ' '
    hour = last_three_characters + rest_of_hour
    hours_korean.append(hour)

max_voltage = 0
max_temp = 0
min_voltage = np.infty
min_temp = np.infty
for index in range(len(hours)):
    voltage_row = volt_2016_df[(volt_2016_df['Date'] == date_power) & (volt_2016_df['Time'] == hours[index])]
    temp_row = temp_2016_df[(temp_2016_df['Date'] == date_weather) & (temp_2016_df['Time'] == hours_korean[index])]
    volts_in_row = voltage_row.iloc[0, 3:].tolist()
    temps_in_row = temp_row.iloc[0, 8:].tolist()
    for voltage in volts_in_row:
        if voltage != 'No Value':
            if (voltage > max_voltage):
                max_voltage = voltage
            if (voltage < min_voltage):
                min_voltage = voltage
    for temp in temps_in_row:
        if temp != 'No Value':
            temp = int(temp)
            if temp > max_temp:
                max_temp = temp
            if temp < min_temp:
                min_temp = temp

print(f'Max Volt: {max_voltage} and Min Volt: {min_voltage}')
print(f'Max Temp: {max_temp} and Min Temp: {min_temp}')
```
